```python
df['Diagnosis'].value_counts().plot(kind='bar')
plt.xlabel('Diagnosis')
plt.ylabel('Count')
plt.title('Diagnosis Count')
plt.show()
```

```python
class_colors = {'M': 'red', 'B': 'green'}
# Create scatter plot
plt.scatter(x=df['Radius (mean)'], y=df['Texture (mean)'], c=df['Diagnosis'].map(class_colors))
plt.xlabel('Radius (mean)')
plt.ylabel('Texture (mean)')

# Create legend for the classes
import matplotlib.patches as mpatches
legend_handles = [mpatches.Patch(color=color, label=label) for label, color in class_colors.items()]
plt.legend(handles=legend_handles)

plt.show()
```
